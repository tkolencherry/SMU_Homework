### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
mask_temp = (weather['Temperature (Fahrenheit)'] < 65) & (weather['Temperature (Fahrenheit)'] >= 55)
mask_wind = weather["Wind Speed (mph)"] < 8
mask_cloud = weather.Visibility < 10

combined_mask = mask_temp & mask_wind & mask_cloud
prime_cities = weather.loc[combined_mask].reset_index(drop = True)
prime_cities.City = prime_cities.City.str.title()
prime_cities
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
name = []
address = []
rating = []

#for each city in the prime cities, bring back the longitude and latitude of the city
for index, row in prime_cities.iterrows():
    params = {
        "radius": 5000,
        "types": "hotel",
        "keyword": "hotel",
        "key": g_key
    }
    lat = prime_cities.Latitude[index]
    long = prime_cities.Longitude[index]

    #search parameters
    params['location'] = f'{lat},{long}'

    #create url and generate json from get request to Google Places API 
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    name_address = requests.get(base_url, params=params).json()

    #had to separate out try/except because for loop was adding in duplicate values for entries 
    #i.e. append name and then if no rating info, append "" for name again in the except block
    try:
        name.append(name_address['results'][0]['name'])
    
    except: 
        name.append("")
        
    try: 
        address.append(name_address['results'][0]['vicinity'])
        
    except:
        address.append("")
        
    try: 
        rating.append(name_address["results"][0]["rating"])
    except:
        rating.append("")
```

```python
#append rows for hotel, address, and rating
prime_cities["Hotel"] = name
prime_cities["Address"] = address
prime_cities["Rating"] = rating

prime_cities
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel}</dd>
<dt>City</dt><dd>{City}, {Country}</dd>
<dt>Address</dt><dd>{Address}</dd>
<dt>Rating</dt><dd>{Rating} Stars</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in prime_cities.iterrows()]
locations = prime_cities[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map
marker = gmaps.marker_layer(locations, info_box_content = hotel_info)
fig.add_layer(marker)

fig
# Display figure
```
